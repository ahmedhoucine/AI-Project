```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your DataFrame
df = pd.read_csv('data.csv')

# Create a crosstab of Education vs Job Title
crosstab = pd.crosstab(df['Education'], df['Job Title'])

# Plot as heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(crosstab, cmap='YlGnBu')
plt.title("Relationship between Education and Job Title")
plt.ylabel("Education")
plt.xlabel("Job Title")
plt.show()
```

```python
import pandas as pd
import numpy as np

# Define function
def parse_numerical(val):
    if pd.isna(val) or val == 'non spécifié':
        return np.nan
    elif '>' in val:
        return float(val[1:]) + 1
    elif '-' in val:
        start, end = val.split('-')
        return (float(start) + float(end)) / 2
    else:
        try:
            return float(val)
        except:
            return np.nan

# Apply the function to create the numeric column
df['Number of Employees Numeric'] = df['Number of Employees'].apply(parse_numerical)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Check for column
print(df.columns)  # should show 'Number of Employees Numeric'

# Plot the distribution
sns.histplot(df['Number of Employees Numeric'].dropna(), bins=20, kde=True)
plt.title("Distribution of Number of Employees")
plt.xlabel("Approximate Number of Employees")
plt.show()
```

```python
df['Number of Candidates Numeric'] = df['Number of Candidates'].apply(parse_numerical)



sns.histplot(df['Number of Candidates Numeric'].dropna(), bins=20, kde=True)
plt.title("Distribution of Number of Candidates")
plt.xlabel("Approximate Number of Candidates")
plt.show()

```
